# while 循環與布林值 (bool)

## while 循環的基本概念

`while` 循環會**持續執行**，只要**條件為真 (True)** 就繼續執行。

```python
while 條件:
    # 執行這些程式碼
```

關鍵點：
- **條件**必須是**布林值 (bool)**：`True` 或 `False`
- 當條件是 `True` 時，執行迴圈內的程式碼
- 當條件變成 `False` 時，跳出迴圈

## 布林值的說明

- `True` = 真/是 → 繼續執行迴圈
- `False` = 假/否 → 停止迴圈


````python
# 範例 1: 使用 while True (無限循環)
# 注意：這會一直執行，需要手動中斷 (Ctrl+C)
count = 0
while True:  # True 永遠是 True，所以會一直執行
    count += 1
    print(f"執行第 {count} 次")
    if count >= 5:  # 設定退出條件
        break  # 強制跳出迴圈
````

````python
# 範例 2: 使用變數作為條件 (會自動變成布林值)
condition = True
count = 0
while condition:  # condition 是 True，所以會執行
    count += 1
    print(f"執行第 {count} 次")
    if count >= 3:
        condition = False  # 改成 False，迴圈就會停止
print("迴圈結束了！")
````

````python
# 範例 3: 使用比較運算 (會回傳 True/False)
count = 0
while count < 5:  # count < 5 會回傳 True 或 False
    count += 1
    print(f"count = {count}")
print("迴圈結束，因為 count 已經 >= 5")
````

````python
# 範例 4: 用 while 實現累加 (類似 for 迴圈)
total = 0
i = 1
while i <= 10:  # 當 i <= 10 是 True 時，繼續執行
    total += i
    i += 1
print(f"1 到 10 的總和 = {total}")
````

````python
# 範例 5: 布林值的自動轉換
# Python 會自動把非布林值轉成 True 或 False

# 以下這些會被當成 False (假的)：
# - False
# - None
# - 0
# - 空字串 ""
# - 空列表 []
# - 空字典 {}

# 其他所有值都是 True

value = 5
while value:  # 5 不是 0，所以是 True
    print(f"value = {value}")
    value -= 1  # 每次減 1，當 value 變成 0 時，就會停止
````

## 總結

**while bool 的運作方式：**

1. `while` 後面的條件會**評估為布林值** (True 或 False)
2. 如果是 `True` → 執行迴圈內的程式碼
3. 如果是 `False` → 跳出迴圈
4. 執行完迴圈內的程式碼後，會**再次檢查條件**
5. 重複這個過程，直到條件變成 `False`

**重要提醒：**
- 一定要確保迴圈條件有機會變成 `False`，否則會變成**無限循環**
- 可以使用 `break` 強制跳出迴圈
- 可以使用 `continue` 跳過這次迴圈，繼續下一次
