```python
# 基本數學運算
a = 10
b = 3

print("=== 基本數學運算 ===")
print(f"a = {a}, b = {b}")
print(f"加法: {a} + {b} = {a + b}")
print(f"減法: {a} - {b} = {a - b}")
print(f"乘法: {a} * {b} = {a * b}")
print(f"除法: {a} / {b} = {a / b}")
print(f"整除: {a} // {b} = {a // b}")
print(f"餘數: {a} % {b} = {a % b}")
print(f"次方: {a} ** {b} = {a ** b}")
```

```python
# 運算優先順序：括號 > 次方 > 乘除 > 加減
result1 = 2 + 3 * 4
result2 = (2 + 3) * 4
result3 = 2 ** 3 * 4

print("\n=== 運算優先順序 ===")
print(f"2 + 3 * 4 = {result1}")      # 14 (先乘後加)
print(f"(2 + 3) * 4 = {result2}")    # 20 (括號優先)
print(f"2 ** 3 * 4 = {result3}")     # 32 (先次方後乘)
```
