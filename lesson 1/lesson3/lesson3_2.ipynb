```python
# 基本數學運算
a = 10
b = 3

print("=== 基本數學運算 ===")
print(f"a = {a}, b = {b}")
print(f"加法: {a} + {b} = {a + b}")
print(f"減法: {a} - {b} = {a - b}")
print(f"乘法: {a} * {b} = {a * b}")
print(f"除法: {a} / {b} = {a / b}")
print(f"整除: {a} // {b} = {a // b}")
print(f"餘數: {a} % {b} = {a % b}")
print(f"次方: {a} ** {b} = {a ** b}")
```
